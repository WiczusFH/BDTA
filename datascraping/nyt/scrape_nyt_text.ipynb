```python
import requests
import json
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import json
import random
from data_access.RedisAccess import RedisAccess
from datetime import datetime
```

Login Required

```python
#log in
browser= webdriver.Chrome()
browser.get("https://myaccount.nytimes.com/auth/login")
ra = RedisAccess(db=1)
```

```python
for i in range(4,12):
    articles = json.load(open(f'nyt_metadata_{i}.json'))
    articles_per_month=min(len(articles),300)
    random_indices = random.sample(range(0, len(articles)),articles_per_month)
    for index in random_indices:
        try:
            """Open page."""
            browser.get(articles[index]["web_url"])
            time.sleep(1)
            """Get Data."""
            html = browser.page_source
            soup = BeautifulSoup(html, 'html.parser')
            article_body = soup.find('section', {'name': 'articleBody'})
            """Get Date"""
            date=datetime.strptime(articles[index]["pub_date"], '%Y-%m-%dT%H:%M:%S%z').strftime('%Y%m%d')
            """Save. """
            ra.persist_raw("nyt",date,article_body.text.strip())
            print("Saved. ")
        except Exception:
            print("Error",articles[index]["web_url"])
```
