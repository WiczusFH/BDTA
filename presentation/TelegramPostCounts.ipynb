```python
import json
import pandas as pd
from matplotlib import pyplot as plt
```

```python
with open('../data_analysis/entry_counts.json', 'r') as f:
    telegram_entry_counts = json.load(f)

entry_counts_df = pd.DataFrame(telegram_entry_counts)
entry_counts_df['date']=pd.to_datetime(entry_counts_df['date'],format='%Y%m%d')
entry_counts_df=entry_counts_df.set_index('date')
```

# Kyiv Independent Post Daily Counts
<small>High post count can be observed at the beginning of Russo-Ukrainian conflict. This is to be expected
due to higher activity at the beginning of the war. The drop in post frequency is likely
due to lower activity but as well as public interest exhaustion.

Some additional outlier post counts can be identified in at the beginning of September 2022 and at the beginning of
october. The spikes potentially could be associated with increased targeting of
civilian infrastructure in these periods.

```python
ax = entry_counts_df.plot(kind='line', title='Post frequency', grid=True)

ax.set_xlabel('Date')
ax.set_ylabel('Posts')
ax.legend(['telegram'])
plt.show()
```
